```python
import numpy as np
from PIL import Image
from legomosaic import BuildConfig, ImageAdjust, full_build, PALETTES

# --- Inputs ------------------------------------------------------------
IMAGE_PATH = "sample_city.png"     # <- your image here
cfg = BuildConfig(
    width_studs=64,                # mosaic width in studs
    adjust=ImageAdjust(
        crop_left=0.0, crop_right=0.0, crop_top=0.0, crop_bottom=0.0,
        saturation=1.1, hue_shift=0, brightness=0, contrast=1.05,
        style="Photo",             # Photo | Poster | Comic | Soft | Pop
    ),
    palette_names=PALETTES["Full"],  # or e.g. PALETTES["Portrait / Warm"]
    layers=2,                      # 2 = plate base + tile art (recommended)
    use_diagonals=True, use_curves=True, use_dots=True, use_wings=True,
    detail=0.75,                   # 0..1 - how eagerly shaped parts are used
    dither=0.0,                    # 0..1 - smoother gradients, noisier fields
)

img = np.array(Image.open(IMAGE_PATH).convert("RGB"))
out = full_build(img, cfg)
print(out.stats)
out.image
```

```python
# Save the outputs
out.image.save("mosaic.png")
with open("mosaic.ldr", "w") as f:
    f.write(out.ldr_text)
print("saved mosaic.png and mosaic.ldr (open the .ldr in BrickLink Studio)")

import pandas as pd
pd.DataFrame(out.bom).head(25)
```
